##### 에어코리아 데이터 수집, 전처리

```python
import pandas as pd
import numpy as np
```

```python
dust2210 = pd.read_excel('./raw/202210_data_past_day.xlsx')
dust2210.drop(axis=0, index=[0], inplace=True) # 두번째 행 삭제
dust2210
```

```python
dust2210.info()
```

```python
# 컬럼명을 영문으로
dust2210.rename(columns={'날짜':'date', 'PM10':'pm10', 'PM2.5':'pm25', 
                         '오 존':'o3', '이산화질소':'no2', '일산화탄소':'co', '아황산가스':'so2'}, inplace=True)
dust2210
```

```python
dust2210.dtypes
```

```python
dust2210 = dust2210.astype({'date':'datetime64[ns]', 'pm10':'int64', 'pm25':'int64', 'o3':'float64', 
                            'no2':'float64', 'co':'float64', 'so2':'float64'})
dust2210.tail()
```

```python
dust2210.dtypes
```

```python
# 날짜를 년, 월, 일로 분해 컬럼 생성
dust2210['year'] = dust2210['date'].dt.year
dust2210['month'] = dust2210['date'].dt.month
dust2210['day'] = dust2210['date'].dt.day
```

```python
dust2210.tail()
```

```python
# 컬럼순서 재정렬
dust2210 = dust2210[['date','year','month','day','so2','co','o3','no2','pm10','pm25']]
dust2210.head()
```

```python
dust2210.isnull().sum()
```

##### 기상청 데이터 확인 및 전처리

```python
weather_2210 = pd.read_excel('./raw/OBS_AWS_TIM_202210_data.xlsx')
weather_2210
```

```python
weather_2210.info()
```

```python
# 지점, 지점명 컬럼 삭제
weather_2210.drop('지점', axis=1, inplace=True)
weather_2210.drop('지점명', axis=1, inplace=True)
```

```python
weather_2210.head()
```

```python
weather_2210.columns = ['date', 'temp', 'wind', 'rain', 'humid']
weather_2210.head()
```

```python
weather_2210['date'] = pd.to_datetime(weather_2210['date']).dt.date
```

```python
weather_2210 = weather_2210.astype({'date':'datetime64[ns]'})
```

```python
weather_2210.info()
```

```python
# 풍속 결측치를 0.0으로
weather_2210['wind'].fillna(0.0, inplace=True)
```

```python
weather_2210.info()
```

```python
# 강수량은 0.1 단위로 측정 --> 0인 데이터는 0.01로 변경
weather_2210['rain'] = weather_2210['rain'].replace([0], 0.01)
weather_2210['rain'].value_counts()
```

```python
weather_2210_new = weather_2210.groupby(weather_2210['date']).mean()
weather_2210_new.head()
```

##### 미세먼지, 날씨 데이터 병합

```python
df_total = pd.merge(dust2210, weather_2210_new, how='inner', left_on='date', right_index=True)
df_total.tail()
```

```python
# 인덱스 재정리
df_total.reset_index(drop=True, inplace=True)
```

```python
df_total.tail()
```

##### 상관관계 및 시각화 

```python
import matplotlib
# from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc, rcParams
import seaborn as sns

import pandas as pd
import numpy as np

font_path = 'C:/Windows/Fonts/HanSantteutDotum-Bold.ttf'
font = font_manager.FontProperties(fname=font_path).get_name() # 'NanumGothic'
rc('font', family=font)
rcParams['axes.unicode_minus'] = False
```

```python
# 모든 요소별 상관관계 확인
df_corr = df_total.corr()
df_corr
```

```python
# 히트맵으로 상관관계 표시
plt.figure(figsize=(15,10))
plt.title('대기정보 요소별 상관관계 히트맵')

sns.heatmap(data=df_corr, annot=True, fmt='.2f', cmap='hot')

plt.show()
```

```python
# 히스토그램 시각화
df_total.hist(bins=20, figsize=(20, 15))
plt.show()
```

```python
# 날짜별 미세먼지 농도 막대그래프
plt.figure(figsize=(15, 5))
plt.title('2022년 10월 미세먼지 일별 농도')
sns.barplot(x='day', y='pm10', data=df_total)
plt.xticks(rotation=0)
plt.show()
```

```python
# 날짜별 초미세먼지 농도 막대그래프
plt.figure(figsize=(15, 5))
plt.title('2022년 10월 초미세먼지 일별 농도')
sns.barplot(x='day', y='pm25', data=df_total)
plt.xticks(rotation=0)
plt.show()
```

```python
# 산점도 Scatterplot
plt.figure(figsize=(15, 6))
plt.title('온도 vs 미세먼지 산점도')
plt.plot(df_total['temp'], df_total['pm10'], marker='o', linestyle='none', alpha=0.5)
plt.xlabel('온도("C)')
plt.ylabel('미세먼지(pm10)')
plt.show()
```

```python
# 산점도 Scatterplot
plt.figure(figsize=(15, 6))
plt.title('미세먼지 vs 초미세먼지')
plt.plot(df_total['pm10'], df_total['pm25'], marker='o', linestyle='none', color='orange', alpha=0.5)
plt.xlabel('미세먼지(pm10)')
plt.ylabel('초미세먼지(pm2.5)')
plt.show()
```

```python
# 산점도 Scatterplot
plt.figure(figsize=(15, 6))
plt.title('바람 vs 이산화질소')
plt.plot(df_total['wind'], df_total['no2'], marker='o', linestyle='none', color='yellow')
plt.xlabel('바람')
plt.ylabel('이산화질소(no2)')
plt.show()
```
